1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"
Real-World Application: Authentication systems

## Real World Application: "Authentication System"

```python
print("-----------Login------------") 
username = "admin"				#given
password = "1234"				#given

print("Enter Username to Login: ")
username1 = input()			#As username and password given is already in "", so there is no need for Typecasting for accepting input

if username ==  username1:
	print("Correct Username")
else:
	print("Invalid Credential")


print("Enter Password : ")
password1 = input()


if password == password1:	
	print("Login Successful! ")
else:
	print("Invalid Credential")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks ≥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
### Real-World Application: "Academic evaluation systems"


```python
marks = [45, 78, 90, 33, 60]				#List of marks
        
pass_cnt = 0
fail_cnt = 0
for m in marks:
    if m >= 50:
        pass_cnt += 1

    else:
        fail_cnt += 1
print("Pass/Fail Analysis")
print(" Total number of pass students: ", pass_cnt)
print("Total number of fail students: ", fail_cnt)


```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list
### Real-World Application: Data preprocessing before analysis



```python
names = [" Alice ", "bob", " CHARLIE "]         #Actual List
clean_list = list()
for name in names:
    clean_list.append(name.strip().lower())
    #append function will add data in clean_list
    #strip function helps removes extra spaces tabs or new lines
    #lower helps making string alphabets convert into lower case
print(clean_list)
    
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters
Real-World Application: Text filtering and validation systems

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    print("Message: ", message)
    print("Length: ", len(message))

    if(len(message) > 10):
        print("Flag:  Message longer than 10 characters!!")
        
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count
Real-World Application: Monitoring and log analysis systems


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_cnt = 0

for log in logs:
    if(log == "ERROR"):
        error_cnt += 1
print("Total Error = ", error_cnt)
```
